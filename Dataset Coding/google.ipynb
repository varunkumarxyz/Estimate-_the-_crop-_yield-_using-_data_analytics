```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crop_data=pd.read_csv("crop_production.csv")
print(crop_data)

print('\n')

print(crop_data.fillna(0))
```

```python
print(crop_data.describe())
```

```python
crop_data = crop_data.dropna()
crop_data
```

```python
crop_data['Yield'] = (crop_data['Production'] / crop_data['Area'])
crop_data.head(5) 
```

```python
sns.barplot(crop_data["State_Name"], crop_data["Production"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(crop_data["State_Name"], crop_data["Yield"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(crop_data["Crop_Year"], crop_data["Production"])
plt.xticks(rotation = 90)
```

```python
sns.barplot(crop_data["Crop_Year"], crop_data["Yield"])
plt.xticks(rotation = 90)
```

```python
sns.lineplot(crop_data["Season"], crop_data["Yield"].mean)
plt.xticks(rotation = 90)
```

```python
sns.lineplot(crop_data["Season"], crop_data["Production"])
plt.xticks(rotation = 90)
```

```python
sns.scatterplot(data=crop_data,x="Area",y="Production")
```

# Conclusions Obtained from the dataset:

Kerela is the largest producer of crop in India.
Production per unit area (Yield ) of Puducherry is maximum.
In the year 2011, crop yield was maximum
